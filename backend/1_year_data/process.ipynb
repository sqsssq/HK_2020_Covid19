```python
import numpy as np
import pandas as pd
import networkx as nx
import csv
```

```python
case_network = pd.read_excel('./2.related_case_no.xlsx')#,sheet_name='individual')
case_network.head()

# 
G = nx.DiGraph()

# related_case no
for index, row in case_network.iterrows():
    if row['Single_case'] == 0:
        G.add_edge(row['Caseno'],row['Relatedcasesno'])

# degree centrality
degree_centrality = nx.degree_centrality(G)
print("degree centrality:")
print(degree_centrality)

between_centrality = nx.betweenness_centrality_source(G, normalized=True)
print("between centrality:")
print(between_centrality)  # the type = dict 

#  out_degree
out_degree = dict(G.out_degree())
print("out_degree")
print(out_degree)

combine_dict = [dict(zip(degree_centrality,v)) for v in zip(degree_centrality.values(), between_centrality.values(), out_degree.values())]
with open('graph_stat.csv','w',newline='') as csvfile:
    file_name = ['CASENO','degree_cent','between_cent','out_degree']
    writer = csv.DictWriter(csvfile,fieldnames=file_name)
    writer.writeheader()
    for d in combine_dict:
        writer.writerow(d)
```

```python
combine_info = [degree_centrality,between_centrality,out_degree]
df = pd.DataFrame(combine_info)
# df.T.head()
df_used = df.T # 
df_used.columns=["degree_center","between_center","out_degree"]
df_used.head()
df_used.to_csv('./case_stat.csv',header=True)
```
